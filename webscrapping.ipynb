```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website = 'https://www.bikewale.com/best-scooters-in-india/?sortField1=modelPopularity&sortOrder1=desc&bodyStyleIds=12%2C5%2C1%2C9%2C13%2C14%2C8%2C11%2C7%2C16%2C15%2C10&fuelType=807&pageNumber=3'
```

```python
response = requests.get(website)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
results = soup.find_all('div', {'class' : 'o-fzptVd o-fzptYr o-fzpimw o-fzpibK o-chzWeu o-cpnuEd o-dsiSgT'})
```

```python
name = []
mileage = []
rating = []
review_count = []
price = []

for result in results:

    # name
    try:
        name.append(result.find('span', {'class':'o-cJrNdO o-bkmzIL o-eqqVmt o-fzpibr'}).get_text().strip())
    except:
        name.append('n/a')

    # mileage
    try:
        mileage.append(result.find('span', {'class':'o-zmksK'}).get_text())
    except:
        mileage.append('n/a')

    # rating
    try:
        rating.append(result.find('p', {'class':'o-eemiLE o-bdcqVx o-cKuOoN o-lIIwF o-cJrNdO'}).get_text())
    except:
        rating.append('n/a')

    # review_count
    try:
        review_count.append(result.find('span', {'class':'o-cwUopP o-fzpimR o-KxopV o-sTQWx o-dThPjR'}).get_text())
    except:
        review_count.append('n/a')

    #price
    try:
        price.append(result.find('span', {'class':'o-cJrNdO o-byFsZJ o-bkmzIL o-bVSleT'}).get_text())
    except:
        price.append('n/a')
```

```python
name
```

```python
# dictionary
EV_list = pd.DataFrame({'Name': name, 'Mileage':mileage,'Rating': rating, 'Review Count': review_count, 'Price': price})
```

```python
EV_list
```

```python
EV_list['Review Count'] = EV_list['Review Count'].apply(lambda x: x.strip('Ratings)').strip('('))
```
